```python
import os
import re
import time
import sys
import pdfquery
import pandas as pd
import numpy as np
from pdfquery.cache import FileCache
```

```python
def f(x):
    pdf = pdfquery.PDFQuery("pdfs/"+x)
    print("Loading",x)
    pdf.load()
    
    # Get List of Items
    cont = pdf.pq(':contains("Near")')
    # Get Date (hopefully)
    dateraw = pdf.pq('LTTextLineHorizontal:contains("Release ")')
    if dateraw.length < 1:
        date = "0000"
    else:
        datet = [c.text for c in dateraw if c.text is not None]#.sub('[^0-9]','', a)#.strip("Release # ")
        print(datet)
        date = re.sub('[^0-9-]','', datet[0])
    print(date)
    # Parse array and join with date
    results = [[date,c.text,x] for c in cont if c.text is not None]

    d = pd.DataFrame(results, columns=['report', 'text','filename'])
    d.replace('', np.nan, inplace=True)
    d.dropna(subset=["text"],inplace=True)
    return d
#     return [[date,c.text] for c in cont if c.text is not None]


# Working
print(f("20170211%20Strike%20Release%20Final.pdf?ver=2017-02-11-074000-027.pdf"))
# Problem
print(f("20161122%20Strike%20Release%20Final.pdf?ver=2017-01-13-131028-640.pdf"))

df = pd.DataFrame(columns=["report","text","filename"])
df = df.append(f("20170211%20Strike%20Release%20Final.pdf?ver=2017-02-11-074000-027.pdf"))
df
```

```python
# This is the main loop
pdfs = os.listdir("./pdfs")
df = pd.DataFrame(columns=["report","text","filename"])
for i,pdf in enumerate(pdfs):
    if pdf.endswith('.pdf'):
        print(i,"of",len(pdfs))
        cont = f(pdf)
        df = df.append(cont)
```
