```python
# https://towardsdatascience.com/how-to-web-scrape-with-python-in-4-minutes-bc49186a8460
import requests
import urllib.request
import os
import re
import time
from bs4 import BeautifulSoup
```

```python
url = 'https://www.inherentresolve.mil/Releases/Strike-Releases/'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, "lxml")
```

```python
links = []
for link in soup.select("a[href*='.pdf']"):
    filename = link['href'].split('href=')[-1]
    filename = filename.replace(" ","%20")
    if(filename[0] == "/"): filename = "https://www.inherentresolve.mil"+filename
    print(filename)
    links.append(filename)
```

```python
delay = .5
print("Download will take approx",len(links)*delay,"seconds, or",(len(links)*delay)/60,"min")
```

```python
for link in links:
    content = requests.get(link)
    if content.status_code==200 and content.headers['content-type']=='application/pdf':
        with open(os.path.join(os.getcwd()+"/pdfs", content.url.rsplit("/",1)[-1]+".pdf"), 'wb') as pdf:
            pdf.write(content.content)
    time.sleep(delay)
```
